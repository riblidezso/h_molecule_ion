```python
import os
import subprocess
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

#Run the program

```python
r_ab_E=[]
m=0
u_i=0
E_i=0
symm="symm"
#symm="antisymm"

E_errs,u_errs,calib=[],[],[]
output=subprocess.check_output(['./errs', str(m), str(u_i), str(E_i), symm ] ,stderr=subprocess.STDOUT)
for line in output.split('\n'):
    print line
u_errs.append(np.loadtxt('u_err.csv',delimiter='\t',skiprows=1))
E_errs.append(np.loadtxt('E_err.csv',delimiter='\t',skiprows=1))
calib.append(np.loadtxt('calib.csv',delimiter='\t',skiprows=1))
```

#Plot the u_err

```python
fig,ax=plt.subplots()
for i in [0]:
    ax.plot(u_errs[i][:,0],u_errs[i][:,1],'x',lw=2,label=r"$Error(\mu)$")
    
ax.axhline(0,c="r",lw=2)
ax.legend(ncol=4)
ax.set_ylim(-10,10)
ax.set_xlim(-10,150)
ax.set_xlabel(r"$\mu$")
ax.set_ylabel(r"$Error$")
fig.set_size_inches(7,5)
fig.savefig('u_err.png',bbox_inches='tight',dpi=200)
```

##Plot E_err

```python
fig,ax=plt.subplots()
for i in [0]:
    ax.plot(E_errs[i][:,0],E_errs[i][:,1],'x',lw=2,label=r"$Error(E)$")
    
ax.axhline(0,c="r",lw=2)
ax.legend(ncol=4)
ax.set_ylim(-0.6,0.6)
ax.set_xlim(-1.5,4)
ax.set_xlabel(r"$E$")
ax.set_ylabel(r"$Error$")
fig.set_size_inches(7,5)
fig.savefig('E_err.png',bbox_inches='tight',dpi=200)
```

##Plot calibration curve

```python
fig,ax=plt.subplots()
for i in [0]:
    ax.plot(calib[i][:,0],calib[i][:,1],'o',lw=2,label=r"$ \mu(\lambda^2)$  calibration")
    
ax.legend(ncol=4)


ax.set_xlabel(r"$\lambda^2$")
ax.set_ylabel(r"$\mu$")
fig.set_size_inches(7,5)
fig.savefig('calib.png',bbox_inches='tight',dpi=200)
```
