```python
import os
import subprocess
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

#Run the program

```python
r_ab_E,calib=[],[]
m=0
u_i=0
E_i=0
#symm="symm"
symm="antisymm"

h_res,r_res,calib=[],[],[]
for u_i in [0,1,2,3]:
    output=subprocess.check_output(['./solve', str(m), str(u_i), str(E_i), symm] ,stderr=subprocess.STDOUT)
    for line in output.split('\n'):
        print line
    h_res.append(np.loadtxt('h_res.csv',delimiter='\t',skiprows=1))
    r_res.append(np.loadtxt('r_res.csv',delimiter='\t',skiprows=1))
    calib.append(np.loadtxt('calib.csv',delimiter='\t',skiprows=1))
```

#Plot the result of H equation

```python
fig,ax=plt.subplots()
for m in [0,1,2,3]:
    ax.plot(h_res[m][:,0],h_res[m][:,1],lw=2,label=r"$\mu_i$ = " +str(m))
    
ax.legend(ncol=4)
ax.set_ylim(-1,0.9)
ax.set_xlabel(r"$\eta$")
ax.set_ylabel("$H(\eta)$")
fig.set_size_inches(7,5)
fig.savefig('u_i_antisymm_plot.png',bbox_inches='tight',dpi=200)
```
