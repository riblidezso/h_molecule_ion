```python
import os
import subprocess
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

#Run the program
- "last last iteration in E calculation" is because of newton-raphson not converging enough close, because of errors and noise, but it is usually no problem now.

```python
r_ab_E,calib=[],[]
m=0
u_i=0
E_i=0
#symm="symm"
symm="antisymm"

output=subprocess.check_output(['./scan_rab', str(m), str(u_i), str(E_i), symm ] ,stderr=subprocess.STDOUT)
for line in output.split('\n'):
    print line
r_ab_E.append(np.loadtxt('r_ab_E.csv',delimiter='\t',skiprows=1))
calib.append(np.loadtxt('calib.csv',delimiter='\t',skiprows=1))
```

```python
m=0
u_i=0
E_i=0
symm="symm"
#symm="antisymm"

output=subprocess.check_output(['./scan_rab', str(m), str(u_i), str(E_i), symm ] ,stderr=subprocess.STDOUT)
for line in output.split('\n'):
    print line
r_ab_E.append(np.loadtxt('r_ab_E.csv',delimiter='\t',skiprows=1))
calib.append(np.loadtxt('calib.csv',delimiter='\t',skiprows=1))
```

#Plot the E_min(r_ab)

```python
fig,ax=plt.subplots()

ax.plot(r_ab_E[0][:,0],r_ab_E[0][:,2],"o",lw=3,label=r"antisymmetric")
ax.plot(r_ab_E[1][:,0],r_ab_E[1][:,2],"o",lw=3,label=r"symmetric")
ax.axhline(-1,c="r",lw=3,label="neutral hydrogen atom")

ax.legend()

ax.set_ylim(-1.3,0)
ax.set_xlim(0.75,10)
ax.set_xlabel(r"$r_{ab}$")
ax.set_ylabel(r"$E$")
fig.set_size_inches(7,5)
fig.savefig('E_r_ab.png',bbox_inches='tight',dpi=200)
```
