#Load results

```python
h_res=np.loadtxt('h_res.csv',delimiter='\t',skiprows=1)
r_res=np.loadtxt('r_res.csv',delimiter='\t',skiprows=1)
```

#Plot the result of H equation

```python
fig,ax=plt.subplots()
ax.plot(h_res[:,0],h_res[:,1],lw=2)
ax.set_xlabel(r"$\eta$")
ax.set_ylabel("H($\eta$)")
fig.set_size_inches(6,4)
```

#Plot the result of R equation

```python
fig,ax=plt.subplots()
ax.plot(r_res[:,0],r_res[:,1],lw=2)
ax.set_xlabel(r"$\xi$")
ax.set_ylabel(r"R($\xi$)")
fig.set_size_inches(6,4)
```
