```python
import os
import subprocess
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

#Run the program

```python
h_res,r_res=[],[]
for i in [0,1,2,3]:
    output=subprocess.check_output(['./solve', '0', '0', str(i), 'symm'] ,stderr=subprocess.STDOUT)
    for line in output.split('\n'):
        print line
    h_res.append(np.loadtxt('h_res.csv',delimiter='\t',skiprows=1))
    r_res.append(np.loadtxt('r_res.csv',delimiter='\t',skiprows=1))
```

#Plot the result of R equation

```python
fig,ax=plt.subplots()
for i in [0,1,2,3]:
    ax.plot(r_res[i][:,0],r_res[i][:,1],lw=2,label=r"$E_i$ = " +str(i))
    
ax.legend(ncol=4)
ax.set_ylim(-0.5,1)
ax.set_xlim(1,12)
ax.set_xlabel(r"$\xi$")
ax.set_ylabel(r"$R(\xi)$")
fig.set_size_inches(7,5)
fig.savefig('E_i_plot.png',bbox_inches='tight',dpi=200)
```
